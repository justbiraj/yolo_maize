```python
import matplotlib.pyplot as plt
import numpy as np

# Defining the relu function
def relu(x):
    return np.maximum(0, x)

# Generating x values
x = np.linspace(-10, 10, 400)
y = relu(x)

# Creating the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = \max(0, x)$') # ReLU function
plt.title('ReLU function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 400)
y = sigmoid(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = \frac{1}{1 + e^{-x}}$')

# Add a grid
plt.grid(True, which='both')

# Add a horizontal line
plt.axhline(y=0, color='k')

# Add a vertical line
plt.axvline(x=0, color='k')

# Add title and labels
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('f(x)')

# Add a legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
# Define the tanh function
def tanh(x):
    return np.tanh(x)

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 400)
y = tanh(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$')

# Add a grid
plt.grid(True, which='both')

# Add a horizontal line
plt.axhline(y=0, color='k')

# Add a vertical line
plt.axvline(x=0, color='k')

# Add title and labels
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('f(x)')

# Add a legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
# Define the gelu function
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

# Create an array of x values from -3 to 3
x = np.linspace(-3, 3, 400)
y = gelu(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = 0.5x(1 + \tanh(\sqrt{\frac{2}{\pi}} (x + 0.044715x^3)))$')

# Add a grid
plt.grid(True, which='both')

# Add a horizontal line
plt.axhline(y=0, color='k')

# Add a vertical line
plt.axvline(x=0, color='k')

# Add title and labels
plt.title('GELU Function')
plt.xlabel('x')
plt.ylabel('f(x)')

# Add a legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
# Define the silu function
def silu(x):
    return x / (1 + np.exp(-x))

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 400)
y = silu(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = x \cdot \frac{1}{1 + e^{-x}}$')

# Add a grid
plt.grid(True, which='both')

# Add a horizontal line
plt.axhline(y=0, color='k')

# Add a vertical line
plt.axvline(x=0, color='k')

# Add title and labels
plt.title('SiLU Function')
plt.xlabel('x')
plt.ylabel('f(x)')

# Add a legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
# Define the relu6 function
def relu6(x):
    return np.minimum(np.maximum(0, x), 6)

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 400)
y = relu6(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = \min(\max(0, x), 6)$')

# Add a grid
plt.grid(True, which='both')

# Add a horizontal line
plt.axhline(y=0, color='k')

# Add a vertical line
plt.axvline(x=0, color='k')

# Add title and labels
plt.title('ReLU6 Function')
plt.xlabel('x')
plt.ylabel('f(x)')

# Add a legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu6(x):
    return np.minimum(np.maximum(x, 0), 6)

def hard_swish(x):
    return x * relu6(x + 3) / 6

x = np.linspace(-6, 6, 1000)
y = hard_swish(x)

# Plotting the hard-swish activation function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Hard-Swish Activation Function')
plt.grid(True)

# Adding the piecewise definition as a text annotation without LaTeX
text = """Hardswish(x) =
    0 if x ≤ -3,
    x if x ≥ +3,
    x * (x + 3) / 6 otherwise"""
plt.text(-6, 1.5, text, fontsize=10, bbox=dict(facecolor='white', alpha=0.7))

plt.show()
```
